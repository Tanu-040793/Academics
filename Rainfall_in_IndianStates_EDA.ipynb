```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\ezbanta\\Desktop\\Learnbay_AI_ML_course\\GitHub Progs\\rainfall in india 1901-2015.csv')
```

```python
df.head()
```

```python
#rename the column

df = df.rename(columns={'SUBDIVISION' : 'States'})
df.tail()
```

```python
#plot the annual rainfall of all states 
plt.figure(figsize=(15,10))
sns.lineplot(x = 'YEAR', y= 'ANNUAL', hue = 'States', data = df)
plt.title('Annual Rainfall of all states',fontsize=15)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Box plot representation of annual rainfall in states
plt.rcParams['figure.figsize']=(20,10)
box = sns.boxplot(x='States', y='ANNUAL', data=df, width=0.8,linewidth=3)
box.set_xlabel('State',fontsize=15)
box.set_ylabel('Annual Rainfall',fontsize=15)
plt.title('Annual Rainfall in States',fontsize=20)
box.tick_params(axis='x',labelsize=10,rotation=90)
box.tick_params(axis='y',labelsize=10,rotation=0)
plt.grid()
```

Median is same for both the states Arunachal pradesh and Coastal Karnataka, but overall annual rainfall received is higher 
in Arunachal Pradesh.

```python
#plot the max rainfall state

plt.figure(figsize=(10,8))
df.groupby(['States','YEAR'])['ANNUAL'].sum().sort_values(ascending=False).plot()

plt.grid()
plt.xlabel('State and Year',fontsize=15)
plt.ylabel('Annual Rainfall',fontsize=15)
plt.title('Highest Rainfall plot',fontsize=15)
```

```python
#Rainfall in each state
plt.figure(figsize=(10,8))
df.groupby(['States'])['ANNUAL'].sum().sort_values(ascending=False).head(20).plot(kind='bar')
plt.ylabel('Total Rainfall')
plt.title('Total Rainfall in all the states',fontsize=15)
plt.grid()
```

Highest Annual rainfall received in Coastal Karnataka among all the states. 

```python
#Mont-wise average rainfall received in all states
plt.figure(figsize=(10,8))
df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG','SEP', 'OCT', 'NOV', 'DEC']].mean().plot(kind= 'bar')
plt.ylabel('Avg Rainfall')
plt.xlabel('Month')
plt.title('Average Rainfall in all the states',fontsize=15)
plt.grid()
```

Average rainfall received in all states is higher in June month.

```python
#seasonwise rainfall in all states
plt.figure(figsize=(10,5))
df[['States', 'Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby('States').mean().sort_values('Jun-Sep').plot.bar(width=0.5,align='center',stacked=True,figsize=(15,7))
plt.ylabel('Rainfall')
plt.xlabel('States')
plt.title('Seasonal Rainfall in all the states',fontsize=15)
plt.grid()
```

Average rainfall is higher in Arunachal pradesh among al the states, and Costal Karnataka has maximum average rainfall from June to Sept.
